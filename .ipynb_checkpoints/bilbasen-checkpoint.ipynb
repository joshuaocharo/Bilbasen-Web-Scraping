# Part 1: Single Page Car Listings (Practice Code: Understanding the Question)

## Imports

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import urllib.parse
```

## HTTP Request

### Store Website in Variable

```python
website = 'https://www.bilbasen.dk/brugt/bil?includeengroscvr=true&pricefrom=0&includeleasing=true&page=1'
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36'}
```

### Get Request

```python
response = requests.get(website, headers=headers)
```

### Status Code

```python
response.status_code
```

### Soup Object

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

### All Listings

```python
results = soup.find_all('div', {'class':'row listing listing-plus bb-listing-clickable'})
```

```python
len(results)
```

```python
results[0]
```

### Target Necessary Data

#### Brand and Model

```python
results[0].find('a').get_text()
```

#### Region

```python
results[0].find('div', {'class': 'col-xs-2 listing-region'}).get_text()
```

#### Km/l

```python
results[0].find('div', {'class': 'col-xs-3 listing-data'}).get_text().strip()
```

#### Price (Pounds)

```python
results[0].find('div', {'class': 'col-xs-3 listing-price'}).get_text()
```

#### Relative Url

```python
#Extract the relative url
relative_url=results[0].find('a',{'class':'listing-heading darkLink'}).get('href')
```

```python
#provide the root element inside the variable
root_url='https://www.bilbasen.dk'
#combine root element and relative url
url_combine = root_url + relative_url
```

```python
url_combine
```

### Put Everything Together Inside a For-Loop

```python
brandandmodel = []
region = []
kml = []
price = []
url=[]

for result in results:
    #brand and model
    try:
        brandandmodel.append(result.find('a').get_text())
    except:
        brandandmodel.append('Not Available')
    
    #region
    try:
        region.append(result.find('div', {'class': 'col-xs-2 listing-region'}).get_text())
    except:
        region.append('Not Available')
    
    #km/l
    try:
        kml.append(result.find('div', {'class': 'col-xs-3 listing-data'}).get_text().strip())
    except:
        kml.append('Not Available')
    
    #price
    try:
        price.append(result.find('div', {'class': 'col-xs-3 listing-price'}).get_text())
    except:
        price.append('Not Available')
    
    #relative url
    try:
        url.append(result.find('a',{'class':'listing-heading darkLink'}).get('href'))
    except:
        url.append('Not Available')
    
```

```python
url
```

### Combine URL

```python
url_combined = []

for link in url:
    url_combined.append(urllib.parse.urljoin(root_url, link))
```

```python
url_combined
```

## Create Pandas Dataframe

```python
all_listings = pd.DataFrame({'Brand and Model' : brandandmodel, 'Region' : region, 'KM/L' : kml, 'Price' : price, 'URL' : url_combined})
```

```python
all_listings
```

#### Data Cleaning

```python
all_listings['KM/L'] = all_listings['KM/L'].apply(lambda x:x.strip('km/l'))
```

```python
all_listings
```

```python
all_listings['Price'] = all_listings['Price'].apply(lambda x:x.strip('kr.').strip('/md.').strip('kr'))
```

```python
all_listings
```

## Output to Excel

```python
all_listings.to_excel('single_page.xlsx', index = False)
```

### Getting URL for Car Listing 1

```python
url_combine
```

```python
cresponse = requests.get(url_combine, headers=headers)
```

```python
csoup = BeautifulSoup(cresponse.content, 'html.parser')
```

### Target Necessary Data

#### Basic Information

##### Car

```python
cnameresults= csoup.find('h1',{'id':'bbVipTitle'}).get('title')
```

##### Basic Details

```python
basicresults = []
for wrapper in csoup.find_all('span', {"class":"value"}):
    basicresults.append(wrapper.get_text(strip=True))
```

##### Detailed Information

```python
detailedresults = []
for wrappr in csoup.find_all('table', {'class':'type1'}):
    detailedresults.append(wrappr.get_text(strip=True))
```

##### Dealer/ Private Contact

```python
contactnameresults = []
contacttelresults= []
contactwebresults= []
for wrapr in csoup.find_all('a', {'class':'dealer-name'}):
    contactnameresults.append(wrapr.get_text(strip=True))
for wraprloc in csoup.find_all('span', {'class':'phone-numbers'}):
    contacttelresults.append(wraprloc.get_text(strip=True))
for wraprweb in csoup.find_all('li', {'class':'website actionLink'}):
    contactwebresults.append(wraprweb.get_text(strip=True))
```

### Create Pandas Dataframe for Car Name, Basic and Detailed Information and Dealer/Private Contact

```python
car_name = pd.DataFrame({'Car Name' : cnameresults}, index=[0])
```

```python
basic_info = pd.DataFrame([basicresults],columns=['Price','Km', 'Km/L', 'First Registration', 'Production Year', 'Model Year'], index=[0])
```

```python
detailed_info = pd.DataFrame({'Deatiled Information' : detailedresults}, index=[0])
```

```python
contact_name = pd.DataFrame({'Dealer/ Private Contact Name' : contactnameresults}, index=[0])
```

```python
contact_tel = pd.DataFrame({'Dealer/ Private Contact Telephone': contacttelresults}, index=[0])
```

```python
contact_web = pd.DataFrame({'Dealer/ Private Contact Website' : contactwebresults}, index=[0])
```

# Part 3: Combining Excel files to Multiple Worksheets

```python
import os
import pandas as pd
cwd = os.path.abspath('') 
files = os.listdir(cwd)  

## Gets all sheets of a given file
df_total = pd.DataFrame()
for file in files:                         # loop through Excel files
    if file.endswith('.xlsx'):
        excel_file = pd.ExcelFile(file)
        sheets = excel_file.sheet_names
        for sheet in sheets:               # loop through sheets inside an Excel file
            df = excel_file.parse(sheet_name = sheet)
            df_total = df_total.append(df)
df_total.to_excel('combined_file.xlsx')
```
